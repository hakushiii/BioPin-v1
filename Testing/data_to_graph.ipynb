```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('final_testing.csv')
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data
```

```python
for x in range(len(data)):
    array = data.iloc[x]['ARRAY']
    array = array.split()
    if '[' in array[0]:
        array[0] = array[0][1:]
    if ']' in array[-1]:
        array[-1] = array[-1][:-1]
    while '' in array:
        array.remove('')
    print(x, array)
    array = [float(string) for string in array]
    
    plt.plot(array)
    plt.savefig(f'finalTest/{x+1}.jpg')
    plt.clf()
```
